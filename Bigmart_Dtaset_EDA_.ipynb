```python
import pandas as pd
from ipywidgets import interact
from ydata_profiling import ProfileReport
```

### Read data set

```python
df=pd.read_excel('/content/bigmart.xlsx')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Item_Weight'].fillna(df['Item_Weight'].mean(),inplace=True)
df['Outlet_Size'].fillna(df['Outlet_Size'].mode()[0],inplace=True)
```

```python
df.isna().sum()
```

```python
df['Item_Fat_Content'].replace({'LF':'Low Fat','low fat':'Low Fat','reg':'Regular'})
df['Item_Fat_Content'].head()
```

```python
@interact
def filter_emp(column='Item_Outlet_Sales',Sales=2097.2700):
    return df[df[column]>Sales]
```

### Ydata_profile

YData Profiling is a powerful tool for data exploration and can help you identify potential issues with your data early on, such as outliers, missing values, and skewed distributions. This information can be used to inform data cleaning and preprocessing steps, as well as feature engineering and model selection.

```python
ydata_profiling=ProfileReport(df)
ydata_profiling.to_file(output_file='ydata_profiling.html')
```
